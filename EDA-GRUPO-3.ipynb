## 3. Transformamos o Limpiamos los datos

Notamos que una columna se encuentra mal escrita, así que la corregimos

```python
df_data.rename(columns={'ï»¿ciudad': 'ciudad'}, inplace=True)
df_data.head()
```

Sabemos que para realizar estadística, los datos de tipo string no nos son muy útiles, por ello es que los transformaremos a valores numéricos

```python
city_map = {'La Serena': 1, 'Valparaiso': 2} #realizamos diccionarios de mapeo
genre_map = {'M': 1, 'F': 2}
adress_map = {'URB': 1, 'RUR': 2}
fam_map = {'MI3': 1, 'MQ3': 2}
estP_map = {'SEP': 1, 'JUN': 2}
trabM_map = {'servicios': 1, 'profesor': 2, 'en_casa': 3, 'salud': 4, 'otro': 5}
trabP_map = {'servicios': 1, 'profesor': 2, 'en_casa': 3, 'salud': 4, 'otro': 5}
razon_map = {'reputacion': 1, 'cercania': 2, 'curriculo': 3, 'otro': 4}
asist_map = {'si': 1, 'no': 2}
internet_map = {'si': 1, 'no': 2}
tutor_map = {'padre': 1, 'madre': 2, 'otro': 3}
apoyoE_map = {'si': 1, 'no': 2}
apoyoF_map = {'si': 1, 'no': 2}
extraP_map = {'si': 1, 'no': 2}
actE_map = {'si': 1, 'no': 2}
qSup_map = {'si': 1, 'no': 2}
pololea_map = {'si': 1, 'no': 2}
nivel_map = {'MUYBAJO': 1, 'BAJO': 2, 'NORMAL': 3, 'ALTO': 4, 'MUYALTO': 5}

df_data['ciudad'] = df_data['ciudad'].map(city_map) #aplicamos los mapas
df_data['genero'] = df_data['genero'].map(genre_map)
df_data['direccion'] = df_data['direccion'].map(adress_map)
df_data['famtam'] = df_data['famtam'].map(fam_map)
df_data['estadoP'] = df_data['estadoP'].map(estP_map)
df_data['trabM'] = df_data['trabM'].map(trabM_map)
df_data['trabP'] = df_data['trabP'].map(trabP_map)
df_data['razon'] = df_data['razon'].map(razon_map)
df_data['asistioNMM'] = df_data['asistioNMM'].map(asist_map)
df_data['internet'] = df_data['internet'].map(internet_map)
df_data['tutor'] = df_data['tutor'].map(tutor_map)
df_data['apoyoextra'] = df_data['apoyoextra'].map(apoyoE_map)
df_data['apoyofam'] = df_data['apoyofam'].map(apoyoF_map)
df_data['extrapaga'] = df_data['extrapaga'].map(extraP_map)
df_data['activextra'] = df_data['activextra'].map(actE_map)
df_data['qsuperior'] = df_data['qsuperior'].map(qSup_map)
df_data['pololea'] = df_data['pololea'].map(pololea_map)
df_data['nivelalco'] = df_data['nivelalco'].map(nivel_map)

df_data
```
